```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dataframe=pd.read_csv("assignment1_dataset.csv")
X = dataframe.customer_number
X = (X - X.mean()) / X.std()
y= dataframe.profit
m=0
b=0
## Length, or number of observations, in our data

alpha= 0.01
Num_iters=1500
```

```python
plt.scatter(X,y)
plt.show()
```

```python
def cost_function(X,y,m, b):
    ## number of training examples
    n = len(y) 
    
    ## Calculate the cost with the given parameters
    J = np.sum((y-(X*m+b))**2)/(2*n)
    
    return J
```

```python
cost_function(X,y,m,b)
```

```python
def gradient_descent(X,y, alpha, num_iters):
    j_history = []
    b=0
    m=0
    n = len(y)

    for i in range(n):    
        yhat= m*X+b
        m = m - (alpha * (-2/n)*sum(X*(y-yhat)))
        b = b -(alpha * (-2/n)*sum(y-yhat))
        cost=cost_function(X,y,m, b)
        j_history.append(cost)
    return m,b,j_history    
```

```python
m,b,j_history=gradient_descent(X ,y ,alpha , Num_iters)
print(m,b,j_history)   
```

```python
plt.plot(j_history, color="red")
```

```python
# Making predictions
Y_pred = m*X + b

plt.scatter(X, y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.show()
```
